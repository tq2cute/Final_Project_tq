```python
#import pandas
import keras
import pandas as pd
import numpy as np
import sqlalchemy  
from sqlalchemy import create_engine
from config import db_password
import psycopg2
```

```python
# Postgres username, password,port and database name
POSTGRES_DBNAME = 'California_Housing' 
POSTGRES_PORT = '5432'
POSTGRES_USERNAME = 'postgres' 
POSTGRES_PASSWORD = 'db_password' 
db_string = f"postgres://postgres:{db_password}@127.0.0.1:5432/California_Housing"
California_housing = create_engine(db_string)
```

```python
#Pull Clean_Cal_House_sale_data from PostgresSQL
Cal_House_sale_data = pd.read_sql_query('''SELECT * FROM public."Clean_Cal_House_sale_data" LIMIT 10;''', California_housing)
Cal_House_sale_data
Cal_House_sale_data.drop(['index'], axis = 1)
```

```python
#Pull Clean_Cal_House_info from PostgresSQL
Cal_House_info = pd.read_sql_query('''SELECT * FROM public."Clean_Cal_House_info" LIMIT 10;''', California_housing)
Cal_House_info
Cal_House_info.drop(['index'], axis = 1)
```

```python
#Pull Zip_Code_County_Maching from PostgresSQL
Zip_Code_County_Maching_df = pd.read_sql_query('''SELECT * FROM public."Zip_Code_County_Maching" LIMIT 10;''', California_housing)
Zip_Code_County_Maching_df
#Zip_Code_County_Maching_df.rename(columns = {'zip_code_tabulation_area':'Zip Code'}, inplace = True)
Zip_Code_County_Maching_df.drop(['index'], axis = 1)
```

```python
# Combine the Cal_House_info, Cal_House_sale_data into a single dataset.
California_Housing_df = pd.merge(Cal_House_info, Cal_House_sale_data, on=["MLS Number", "MLS Number"])
California_Housing_df.head()
#drop unecessary index
California_Housing_df['Zip Code'] = (California_Housing_df['Zip Code']).apply(int)
California_Housing_df.drop(['index_x', 'index_y'], axis = 1) 
```

```python
California_Housing_df.dtypes
```

# Preparing Dataframe for Machine Learning Model Outcome

```python
# Merge_California_Housing_df = pd.merge(California_Housing_df, Zip_Code_County_Maching_df, on=["Zip Code"], how='outer')
Merge_California_Housing_df = pd.concat([California_Housing_df, Zip_Code_County_Maching_df], axis=1, join="inner")
```

```python
Merge_California_Housing_df
Merge_California_Housing_df = Merge_California_Housing_df.drop(['index_x', 'index_y', 'index', 'local_health_jurisdiction'], axis = 1)
```

```python

Merge_California_Housing_df = Merge_California_Housing_df.dropna()
Merge_California_Housing_df
```

```python
X = Merge_California_Housing_df.drop(columns=['Sale Price','Sale Date','Listing Date','Street Address','county','MLS Number', 'Close Date', 'zip_code_tabulation_area'])
Y = Merge_California_Housing_df[['Sale Price']]
```

```python
print(X)
```

```python
print(Y)
```

```python
!pip install keras
```

```python
model = keras.models.Sequential()

model.add(keras.layers.Dense(10, activation='relu', input_shape=(10,)))
model.add(keras.layers.Dense(10, activation='relu'))
model.add(keras.layers.Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(X, Y, epochs=30, callbacks=[keras.callbacks.EarlyStopping(patience=3)])
```

```python
test_data = np.array([94941,17,1750000,43,2215,31080.0,4,1978,2021,2021])
print(model.predict(test_data.reshape(1,10), batch_size=1))
```
